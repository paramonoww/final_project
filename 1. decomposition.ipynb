# Телеком — Определение неэффективных операторов

### Задача:

Помогите «Нупозвони» найти самых неэффективных операторов. Признаки низкой эффективности:

- *много пропущенных входящих вызовов*
- *долгое ожидания ответа при входящих внешних вызовах*
- *мало исходящих вывозов — в случае тех колл-центров, что специализируются на обзвонах, а не обработке входящих звонков*

### Этапы исследования:

1. ***Подготовка к анализу данных***
* *Импорт нужных библиотек*
* *Загрузка датасетов*
* *Краткий взгляд на данные*
* *Описание данных*
* *Промежуточные выводы*

2. ***Предобработка данных***
* *Названия столбцов*
* *Типы данных в столбцах*
* *Дубликаты*
* *Пропущенные значения*
* *Аномальные значения*

3. ***Исследовательский анализ данных***
* *Посмотреть распределения в данных и построить boxplots и histplots*
* *Подробно исследовать звонки и их типы*
* *Рассмотреть каждый тарифный план в отдельности и найти слабые и сильные места*

4. ***Определение неэффективных операторов***
* *Найдя слабые и сильные места - определить неэффективных операторов. Будем сранивать тарифный план, длительность звонков, кол-во звонков, соотношения внутренних и внешних звонков, входящих и исходящих, кол-во вызовов по дням, затем сравненим тарифы между собой*
* *Предложить варианты по оптимизации проблемы*


5. ***Проверка статистических гипотез***
* *Средняя длительность звонков тарифа A больше, чем у тарифа B*
* *Средняя длительность звонков тарифа А больше, чем у тарифа C*
* *Кол-во звонков у тарифа А больше, чем у тарифа B*
